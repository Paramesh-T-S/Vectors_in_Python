```python
#TO plot all the vectors in one go
def plot_vectors(vecs):
    colors = ['r','b','g','y','o']
    i=0
    for vec in vecs:
        plt.quiver(vec[0],vec[1],vec[2],vec[3], scale_units='xy',angles='xy', scale = 1,color =colors[i%len(colors)])
        i=i+1
    plt.xlim(-10,10)
    plt.ylim(-10,10)
    plt.show()
```

```python
plot_vectors([(0,0,3,4),(0,0,-3,4),(0,0,-4,-7),(0,0,5,-6)])
```

```python
vecs = [(0,0,3,4),(0,0,-3,4),(0,0,-4,-7),(0,0,5,-6)]
```

```python
plot_vectors([vecs[0],vecs[3]])
```

```python
vecs[0]+vecs[3] #It actually concatenates but we need to add
```

```python
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-4,-7]),np.asarray([0,0,5,-6])] #So converting list in to array to perform arithmetic operation
```

```python
vecs[0]+vecs[3]
```

```python
plot_vectors([vecs[0],vecs[3],vecs[0]+vecs[3]]) #Vector addition
```

```python
plot_vectors([vecs[0],vecs[3],vecs[0]-vecs[3]]) #Vector subtraction
```

```python
#Dot product in vector
vecs = [np.asarray([0,0,3,4]),np.asarray([0,0,-3,4]),np.asarray([0,0,-4,-7]),np.asarray([0,0,5,-6])]
plot_vectors(vecs)
```

```python
#For dot product we need only the destination
a = np.asarray([9, 2])
b = np.asarray([-5, 4])
```

A  B = |A| * |B| * cos() =  A1B1 + ... + AnBn

```python
a_dot_b = np.dot(a, b)
```

```python
print(a_dot_b)
```

proj a = 	a  b/|b|

```python
proj_a = np.dot(a, b)/np.linalg.norm(b)
```

```python
print(proj_a)    #Magnitude
```

To Find Direction :

a = (a.b/|b|)*b

```python
vec_a_b = (proj_a/np.linalg.norm(b))*b
```
